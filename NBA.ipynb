#### Tratamiento de los Datos

##### Trabajando con datos nulos

```python
df_0 = df_nba.copy()
df_0.head()
```

```python
#Eliminado fila 457 suyos todos sus valores son nulos
df_file = df_0.drop(457)
df_file
```

```python
#Eliminando columna College
df_drop_colege = df_file.drop(columns = "College")
df_drop_colege
```

##### Tratamiento en la columna salary

```python
df_1 = df_drop_colege.copy()
df_1
```

```python
#Eliminando dato nulo unico en Salary
df_salary_drop = df_1.dropna(subset= "Salary")
df_salary_drop
```

```python
df_salary_drop["Salary"] = df_salary_drop["Salary"].astype(int)
df_salary_drop
```

```python
#Creando nueva columna prteneciente a Salary
df_salary_drop["Salary_format"] = df_salary_drop["Salary"]
df_salary_drop
```

```python
df_salary_2 = df_salary_drop.copy()
```

```python
#Creando una funcion para formatear los numeros en salary
def formatear_millones(x):
    return '{:,.0f}'.format(x).replace(',', '.')
```

```python
#Aplicando la funcion a la columna
df_salary_2["Salary_format"] = df_salary_2["Salary_format"].apply(formatear_millones)
df_salary_2
```

#### Formateo listo

##### Tratamiento en columna [Age, Numer, Weight]
* the next step will be choose specific columns that have numerical values in float and the task will be convert to int

```python
df_2 = df_salary_2.copy()
df_2
```

```python
#Removing specific columns with the method .drop
df_2_drop = df_2.drop(columns = ["Number","Age","Weight"])
df_2_drop
```

```python
#Selecting the columns that we are gonna to work 
df_2_chose = df_2[["Number", "Age", "Weight"]]
df_2_chose
```

```python
#Using the method .astype to change the type data from float to int
df_2_integer = df_2_chose[["Number","Age","Weight"]].astype(int)
df_2_integer
```

```python
#Joining both df 
df_concat = pd.concat([
    df_2_integer,
    df_2_drop,
],axis=1, join='inner')
```

```python
#The result
df_concat
```
