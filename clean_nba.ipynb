##### Using the method replace in the column Height

```python
df_7 = df_6.copy()
df_7.head()
```

```python
df_7['Height'].unique()
```

```python
df_height = df_7.copy()
```

```python
#Replacing data in height
df_height['Height'] = df_height['Height'].replace('6-2',188)
df_height['Height'] = df_height['Height'].replace('6-6',201)
df_height['Height'] = df_height['Height'].replace('6-5',198)
df_height['Height'] = df_height['Height'].replace('6-10',185)
df_height['Height'] = df_height['Height'].replace('6-9',210)
df_height['Height'] = df_height['Height'].replace('6-8',207)
df_height['Height'] = df_height['Height'].replace('7-0',213)
df_height['Height'] = df_height['Height'].replace('6-4',195)
df_height['Height'] = df_height['Height'].replace('5-9',179)
df_height['Height'] = df_height['Height'].replace('6-7',204)
df_height['Height'] = df_height['Height'].replace('6-3',192)
df_height['Height'] = df_height['Height'].replace('5-11',180)
df_height['Height'] = df_height['Height'].replace('6-11',186)
df_height['Height'] = df_height['Height'].replace('7-3',222)
df_height['Height'] = df_height['Height'].replace('6-0',182)
df_height['Height'] = df_height['Height'].replace('7-2',219)
df_height['Height'] = df_height['Height'].replace('6-1',185)
df_height['Height'] = df_height['Height'].replace('7-1',216)
```
